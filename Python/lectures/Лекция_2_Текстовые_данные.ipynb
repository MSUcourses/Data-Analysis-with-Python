## Функции input и print.
Начнём создавать программу.
Для начала нужно осознать, что результат любых вычислений — это *объект*, который можно запомнить.
Чтобы что-то запомнить, нужно это *назвать*, а для этого используется *оператор присваивания*, который в Python связывает имя с объектом. 

```python
name = input("Ваше имя: ")
```

Теперь мы сможем обращаться к объекту по имени `name`, например, чтобы его распечатать:

```python
print(name)
```

Слова `input` и `print` — это *функции*, которые мы вызываем. В скобках через запятую перечисляются параметры, которые передаются функции. Функция содержит некий скрытый от нас алгоритм, который запускается в момент её вызова. Результат работы алгоритма возвращается и может быть сохранён под каким-то именем, например, `name`, как показано выше.

Некоторые функции не возвращают ничего. Если это "ничего" попытаться запомнить, то оно окажется объектом `None`:

```python
something = print()
```

Кстати, обратите внимание, что здесь функции `print` не было передано ни одного параметра, но круглые скобки всё равно стоят. Без них алгоритм функции не будет запущен, т.е. функция не будет вызвана.

```python
print(something)
```

Можно выводить значения любых выражений. Например, допустима запись `print(3 + 5 * 2)`.
Можно выводить значения нескольких выражений, перечислив их через запятую:

```python
a = 1
b = 2
print(a, '+', b, '=', a + b)
```

В данном случае будет напечатан текст `1 + 2 = 3`: сначала выводится значение `a`, затем строка из знака “+”, затем значение `b`, затем строка из знака “=”, наконец, значение суммы `a + b`.

При выводе значения разделяются одним пробелом. Это поведение можно изменить — разделять выводимые значения любым другим символом и даже любой другой строкой, выводить их в отдельных строках или не разделять никак. Для этого алгоритму функции `print` нужно передать специальный **именованный параметр** `sep` (sep — аббревиатура от слова separator, т.е. разделитель). По умолчанию разделитель равен строке из одного пробела, и именно поэтому между значениями выводится пробел. Передадим параметр sep, равный строке ':':

```python
print(a, b, sep=':')
```

Чтобы совсем убрать разделитель при выводе нужно передать параметр `sep`, равный пустой строке `''`:

```python
print(a, '+', b, '=', a + b, sep='')
```

Можно как значение параметра `sep` передать строку, состоящую из специального символа новой строки `'\n'`:

```python
print(a, b, sep='\n')
```

Символ обратного слэша в литералах строк — это начало *спецсимвола*. Чаще всего употребляется символ новой строки `'\n'` и символ *табуляции* `'\t'` (широкий пробел с выравниванием). А для того, чтобы вставить в строку сам символ обратного слэша, нужно повторить его два раза: `'\\'`.

Ещё один именованный параметр функции `print` является параметр `end`. Он обозначает строку, которая выводится *после* вывода всех значений. По умолчанию параметр `end` равен `'\n'`, поэтому каждый следующий вывод начинается с новой строки. Но и этот параметр можно изменить:

```python
print(a, end = '')
print(b, end = '')
print(a + b, end = '')
```

## Литералы строк. Операции + и * со строками
Текстовые строки являются одним из основных объектов. *Литерал строки* в Python может ограничиваться апострофами, кавычками, а в некоторых случаях даже тремя апострофами или тремя кавычками — и это не меняет тип объекта. Внутри апострофов знак кавычки является просто символом, а внутри кавычек апостроф теряет свой синтаксический смысл. Например, `"I'm a student."` удобнее написать в кавычках, а `'Я зарегистрировался в системе "Истина"'` в апострофах. Если внутри литерала в кавычках нужно записать кавычку или в апострофах апостроф, то его можно экранировать обратной косой чертой (back slash): `"Фильм \"Д'Артаньян и три мушкетёра\"."`.

```python
print("I'm a student.")
print('Я зарегистрировался в системе "Истина"')
print("Фильм \"Д'Артаньян и три мушкетёра\".")
```

## Ошибки и как к ним правильно относиться.
Увы, иногда вы будете писать синтаксическую чушь, а интерпретатор будет "ругаться" на вас сообщениями об ошибках.

Даже в программе, которую пишет профессионал, всегда есть ошибки, хотя не всегда они всплывают наружу. Иногда поиск и устранение одной ошибки требует от профессионала нескольких суток работы...

Запомните, **явная ошибка намного лучше, чем скрытая**, а значит те ошибки, на которые интерпретатор "ругается", легко осознать и устранить. ПОЛЮБИТЕ сообщения об ошибках! Научитесь их *внимательно читать*.

# Память программы
Программа, запущенная на исполнение, называется **вычислительным процессом**. У каждого вычислительного процесса на компьютере есть ограниченный запас ресурсов, который распределяет **операционная система**. Один из самых важных ресурсов — это *оперативная память* (RAM) компьютера. Вычислительный процесс может запрашивать дополнительную память у операционной системы, но если он не будет её освобождать после использования, ему просто перестанут её давать.

Для того, чтобы меньше думать о выделении и освобождении памяти, в Python есть **сборщик мусора**. Это автоматический помощник, который выбрасывает всё, на что программа больше не ссылается. Такая система называется "Автоматическая сборка мусора".

*Имя — это ссылка на объект. Назвать — значит запомнить.*

Если объект не назвать, он будет временным, но может успеть сослужить свою службу, поучаствовав в вычислении выражения.

# Текстовые файлы: как открыть и прочитать. Как и зачем закрывать.
Работа с текстовыми файлами почти так же легка, как с объектами строки.

Для начала создадим файл:


```python
file = open('my.txt', 'w')
file.writelines(["Ехал грека\n", "Через реку,\n",
                 "Видит грека -\n", "В реке рак.\n"])
file.close()
```

лекция 1_17.png

```python
file = open('my.txt')
for line in file:
    print(line)
```

лекция 1_11.png

Поскольку файл нам больше не нужен, можно не называть его словом `file`, а сделать временным.

Также мы видим, что при распечатке появились двойные переводы строки. Это потому, что `line` содержит переводы строк, содержавшиеся в файле. Давайте уберём, "разденем" их методом объекта строки `strip()`:

```python
for line in open('my.txt'):
    print(line.strip())
```

Для "вежливости" по отношению к операционной системе и пользователю не хватает только закрыть файл по окончании считывания. Иначе файл будет заблокирован вычислительным процессом, и нельзя будет вытащить флешку с файлом или просто удалить файл.

```python
file = open('my.txt')
for line in file:
    print(line.strip())
file.close()
```

Слишком легко забыть о закрытии файла. Чтобы этого не случилось, используем **менеджер контекста with**, который автоматически закроет файл, когда закончится вложенная секция:

```python
with open('my.txt') as file:
    for line in file:
        print(line.strip())
```

лекция 1_14.png

При этом каждую строчку файла можно обрабатывать посимвольно:

```python
with open('my.txt') as file:
    for line in file:
        for symbol in line.strip():
            print(symbol, ord(symbol), end=', ')
        print()
```
